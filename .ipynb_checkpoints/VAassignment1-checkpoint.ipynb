```python
#import library 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
from pandas.plotting import scatter_matrix 
```

```python
#import dataset 
bank_data = 'data/bank.txt'
df = pd.read_csv('data/bank.txt', delimiter='\t')
# print out dataset
df
```

```python
#Check if imported dataset is match with the original dataset using Python for loop
def check_value(): #create nested loop in order to check every data in 2 files 
    for i in bank_data:
        for j in df:
            if i != j: #check if values in first file is equal to the values in the second file
                print("Match")
            else:
                print("Invalid Data")
check_value()
print(df.dtypes) #check the type of the columns
```

```python
#Imported data is equivalent to the orginal dataset
```

```python
#Deal with Upper/lower case:
def get_upper():
    for k in df: #creating a loop to run automatically through the whole data file
        if df[k].dtype == "object": #condition: check if this value is a string, not a interger or float
            df[k] = df[k].str.upper() #if the value is string, return the uppercase string 
get_upper()
```

```python
#Deal With Extra White Spaces:
def remove_whitespace():
    for j in df: #creating a loop to run automatically through the whole data file
        if df[j].dtype == "object": #condition: check if this value is a string, not a interger or float
            df[j] = df[j].str.strip() #return the same string without leading or trailing extra whitespaces
            
remove_whitespace()
```

```python
#Deal with Data Entry Errors (Typos):
#Find the errors
def check_typos():
    for i in df: #creating a loop to run automatically through the whole data file
        if df[i].dtype == "object": #check the data type, if it is a string, count it.
            print(df[i].value_counts()) #return the number of unique value 
check_typos()
```

```python
#From the printed results, there are a number of typo errors in the "job" and "day_of_week" column
def fix_typos():
    df['job'] = df['job'].apply(lambda x: "SERVICES" if x == "SERVICESS" else x)
    df['job'] = df['job'].apply(lambda x: "ENTREPRENEUR" if x == "ENTREPRENEURS" else x)
    df['job'] = df['job'].apply(lambda x: "BLUE-COLLAR" if x == "BLUECOLLAR" else x)
    df['day_of_week'] = df['day_of_week'].apply(lambda x: "MON" if x == "MONDAY" else x)
    df['day_of_week'] = df['day_of_week'].apply(lambda x: "FRI" if x == "FRIDAY" else x)
fix_typos()
check_typos() #print the fixed result and check if it there any mistake that has not been dealed with
```
