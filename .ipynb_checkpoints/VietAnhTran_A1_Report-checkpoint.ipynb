<b> 1. Data Import </b>

```python
#import library 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
from pandas.plotting import scatter_matrix 
```

```python
#import dataset 
bank_data = 'data/bank.txt'
df = pd.read_csv('data/bank.txt', delimiter='\t')
# print out dataset
df
```

Imported data is equivalent to the original data source

<b> 1.3: Upper and lower case <b>


This task convert all text values into upper case letter to treat as one

```python
#Deal with Upper/lower case:
def convert_upper():
    for i in df: # Loop through imported data
        if df[i].dtype == "object": # Get all text value
            df[i] = df[i].str.upper() # Convert string value into upper case letters
convert_upper()
```

<b> 1.4: Typos error

This task identify whether typo mistakes exist in the value and fill incorrect values

```python
#Find the errors
def check_typos():
    for i in df: # Loop through the dataframe
        if df[i].dtype == "object": # Get object type values
            print(df[i].value_counts()) # Return frequency of object values
check_typos()
```

```python
#From the printed results, there are a number of typo errors in the "job" and "day_of_week" column
def fix_typos():
    df['job'] = df['job'].apply(lambda x: "SERVICES" if x == "SERVICESS" else x)
    df['job'] = df['job'].apply(lambda x: "ENTREPRENEUR" if x == "ENTREPRENEURS" else x)
    df['job'] = df['job'].apply(lambda x: "BLUE-COLLAR" if x == "BLUECOLLAR" else x)
    df['day_of_week'] = df['day_of_week'].apply(lambda x: "MON" if x == "MONDAY" else x)
    df['day_of_week'] = df['day_of_week'].apply(lambda x: "FRI" if x == "FRIDAY" else x)
    df["job"] = df['job'].str.replace('[^\w\s]','') #remove punctuation
fix_typos()
check_typos() #print the fixed result and check if it there any mistake that has not been dealed with
```

<b> 1.5: Remove extra-whitespaces

This task identify redundant whitespaces in the data and remove them 

```python
#Deal With Extra White Spaces:
def remove_whitespace():
    for i in df: # Loop through the data
        if df[i].dtype == "object": # Get object type values
            df[i] = df[i].str.strip() # Return the string without redundant whitespaces
            
remove_whitespace()
```

<b> 1.7: Missing values

In this task, missing values will be identify and fill or drop depends on the importance and the shape of the whole column

```python
# check na
df.isna().sum()
```

```python
df.describe() #get an basic stats of the dataset
```

```python
# 'duration' column has the minumum value of 0 which is not realistic
# ignore 'duration' column to avoid potential leakage if we need this model to be trained 
del df['duration']
```

```python
# Fill null values in housing coloumn with UNKNOWN 
df['housing'] = df['housing'].fillna('UNKNOWN')
```

```python
# Fill null values in campaign, cons.price.idx, cons.conf.idx columns with median value 
df['campaign'] = df['campaign'].fillna(df['campaign'].median())
df['cons.price.idx'] = df['cons.price.idx'].fillna(df['cons.price.idx'].median())
df['cons.conf.idx'] = df['cons.conf.idx'].fillna(df['cons.conf.idx'].median())
```

```python
# Fill null values in emp.var.rate and euribor3m columns with mean value since the data is skewed
df['emp.var.rate'] = df['emp.var.rate'].fillna(df['emp.var.rate'].mean())
df['euribor3m'] = df['euribor3m'].fillna(df['euribor3m'].mean())
```

```python
# Check if is there any null values that have not been treated
df.isna().sum()
```

<b> 1.8: Categorical data

This task will identify categorical type in the dataset and convert them to suitable type

```python
df.info() # Check the current data type of the set
```

Loan, housing and default column is defined as yes or no state But for the sake of treating the "UNKOWN" values, convert every object Dtype into category type instead of bool

```python
for col in ['job', 'marital', 'education', 'default','housing','loan','contact','month','day_of_week','poutcome']:
    df[col] = df[col].astype('category') #loop through objects columns, change them into category data type
```

```python
df['y'] = df['y'].astype('bool') #Encode "y" column into boolean data type since it only represent true or false state
```

```python
# Check the results
df.info()
# Encode 
df['job1']= df['job'].cat.codes
df['marital1']= df['marital'].cat.codes
df['education1']=df['education'].cat.codes
df['housing1']=df['housing'].cat.codes
df['loan1']=df['loan'].cat.codes
df['contact1']=df['contact'].cat.codes
df['month1']=df['month'].cat.codes
df['day_of_week1']=df['day_of_week'].cat.codes
df['poutcome1']=df['poutcome'].cat.codes
```

<h3> 2.2: Relationship </h3>

<h4> a) Age and Job 

Initial hypothesis: People who is younger in age (below 25 years old) tends to be single and people who are older in age are either married or divorced

```python
#set the plot area
figure = plt.figure(figsize=(25,10)) 
# define axis 
ax = figure.gca()    
df[['age','marital']].boxplot(by = 'marital', ax = ax) 

#set title, x lable and y lable
ax.set_title('Realationship between age and marital status') 
ax.set_xlabel('Marital status')
ax.set_ylabel('Age')
```

Genrally, regardless of the "unkown" value of data, result present the similar idea compared to the initial hypothesis. From the graph, it is clear that the range of age in single category are much lower than the one of divorced and married (18-50 compared to 21-75). 75% of people who are single are at the age from below 38 years old while this indicator of the other two age group are below 52 years and 50 years old. However, looking at the outlier of the result, there are people from 51-75 years old who have not been in marital relationship. 

<h4>b) Loan and Consumer Price Index

Initial hypothesis: People who has personal loan tends to spend less money since they have to set aside an amount of money to pay off their debt

```python
#set the plot area
figure = plt.figure(figsize=(25,10)) 
# define axis 
ax = figure.gca()    
df[['loan','cons.price.idx']].boxplot(by = 'loan', ax = ax) 

#set title, x lable and y lable
ax.set_title('Realationship between loan and consumer price index') 
ax.set_xlabel('Loan status ')
ax.set_ylabel('Consumer Price Index')
```

On the contrary to the initial hypothesis, the difference in term of personal spending between people who have or do not have loan is not significant.From the graph, 25% and 75% indicator are almost identical (with the diffrence of more than 1/100). The minor difference when taking a closer look at the result is that on average, people who are in personal loan program have lower consumer price index compared to the one without (<93.4 and <93.8 respectively).

<h4>c) Consumer Confidence Index vs Consumer Price Index

Initial hypothesis: People who has highe consumer confidence index (CCI) tends to spend more monthly which increase their consumer price index (CPI)

```python
#set the plot area
figure = plt.figure(figsize=(18,10)) 
# define axis 
ax = figure.gca()    
sns.histplot(data=df, x=df['cons.conf.idx'], y=df['cons.price.idx'])
#set title, x lable and y lable
# plt.legend(loc = "upper left", bbox_to_anchor = (1, 1) )
ax.set_title('Relationship between consumer confidence index and consumer price index') 
ax.set_xlabel('CCI')
ax.set_ylabel('CPI')
```

As expected, the analysis results in a down slope where people who have higher CPI tends to have higher CCI and vice versa. However, there is a few significant of people who is in the highest range in terms of CCI who have low consumer price index.
