```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bank_df = pd.read_csv('bank.csv', sep='\t')
```

```python
bank_df.info()
```

### check data types

```python
bank_df.dtypes
```

### typos

```python
bank_df.head()
```

```python
columns = bank_df.columns[bank_df.dtypes == 'object'].tolist()
```

```python
bank_df.job.str.replace(r'[^a-zA-Z0-9 ]', " ")
```

```python
for column in columns:
    bank_df[column] = bank_df[column].str.replace(r'[^a-zA-Z0-9 ]', " ")
```

```python
bank_df
```

### check extra whitespace: using strip()

```python
# https://www.w3schools.com/python/ref_string_strip.asp
for column in columns:
    bank_df[column] = bank_df[column].str.strip()
```

### change columns name

```python
bank_df.info()
```

```python
bank_df = bank_df.rename(columns={"emp.var.rate": "emp_var_rate", "cons.price.idx": "cons_price_idx", "cons.conf.idx": "cons_conf_idx", "nr.employed": "num_employees" })

```

### missing values

```python
bank_df.info()
```

```python
bank_df_2 = bank_df.dropna()
```

```python
bank_df_2.info()
# if we drop all nan values, we will lose many information -> replace by mean or median which have lighter affect.
```

```python
a = bank_df[bank_df['duration'].isnull()]
```

```python
a
```

```python
a.duration
```

<p> according to the requirement, if duration = 0 -> y: "no". so in this case, we will replace the NaN values by 0 </p>


```python
bank_df['duration'] = bank_df['duration'].fillna(0)
```

```python
bank_df.info()
```

```python
bank_df['campaign'] = bank_df['campaign'].fillna(bank_df['campaign'].mean())
```

```python
bank_df['housing'] = bank_df['housing'].fillna('unknown')
```

```python
bank_df['emp_var_rate'] = bank_df['emp_var_rate'].fillna(bank_df['emp_var_rate'].mean())
```

```python
bank_df['cons_price_idx'] = bank_df['cons_price_idx'].fillna(bank_df['cons_price_idx'].mean())
```

```python
bank_df['cons_conf_idx'] = bank_df['cons_conf_idx'].fillna(bank_df['cons_conf_idx'].mean())
```

```python
bank_df['euribor3m'] = bank_df['euribor3m'].fillna(bank_df['euribor3m'].mean())
```
